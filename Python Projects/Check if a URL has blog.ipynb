```python
# Specify the path to your Excel file
excel_file_path = r"D:\Ahrefs\Check if theres blog.xlsx"

# Load the Excel file into a pandas DataFrame
df = pd.read_excel(excel_file_path, sheet_name="Sheet1")
```

```python
df2 = df["Harvested Landing Page"]
df3 = df2.iloc[1:1000]
```

```python
import concurrent.futures
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin

def find_blog_links(base_url):
    try:
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.0.0 Safari/537.36"
        }

        response = requests.get(base_url, headers=headers)

        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find all links on the page
            links = [a['href'] for a in soup.find_all('a', href=True)]

            # Filter links that contain "blog" in their URLs
            blog_links = [urljoin(base_url, link) for link in links if 'blog' in link.lower()]

            return base_url, blog_links

        return base_url, []  # No links found or an error occurred

    except Exception as e:
        return base_url, []  # An error occurred

def find_blog_links_parallel(urls, num_workers=20):
    blog_links_list = []

    with concurrent.futures.ThreadPoolExecutor(max_workers=num_workers) as executor:
        # Use the map method to parallelize finding blog links for multiple URLs
        blog_links_list = list(executor.map(find_blog_links, urls))

    return blog_links_list

if __name__ == "__main__":
    # List of base URLs you want to crawl (replace with your own list)
    base_urls = df3

    # Call the function to find blog links in parallel
    blog_links_list = find_blog_links_parallel(base_urls)

    # Print the results
    for base_url, blog_links in blog_links_list:
        if blog_links:
            print(f"Base URL: {base_url}, Blog Links: {blog_links}")
        else:
            print(f"No blog links found for {base_url}.")
```

```python
import csv

with open("blog_links.csv", "w", newline="") as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["Base URL", "Blog Links"])

    for base_url, blog_links in blog_links_list:
        csv_writer.writerow([base_url, ", ".join(blog_links)])

print("Results saved to 'blog_links.csv'")
```
